```python
url="https://www.google.com/maps/place/Vadodara,+Gujarat/@22.3220425,73.0330003,11z/data=!3m1!4b1!4m5!3m4!1s0x395fc8ab91a3ddab:0xac39d3bfe1473fb8!8m2!3d22.3071588!4d73.1812187"
resp=requests.request(method='GET',url=url)
```

```python
html_content=soup_parser.html.contents[1]
html_content
```

```python
_script = html_content.find_all("script")[7]
_script
```

```python
matches=re.findall("(\d+\.\d{7})",_script.text)
print(matches)    

#'\d' used to match one decimal digit (from 0 to 9)
#character '+' specifies an expression to its left for one or more times.
#'\d{7}' used to match a particular number of occurrences defined in the curly bracket for its left string.(7 digits after decimal point)
#Metacharacter '.' is used to match any single character in a string except the line terminator, i.e. /n.
```
