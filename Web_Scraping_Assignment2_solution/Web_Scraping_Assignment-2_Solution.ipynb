# Answer-1) Scraping Data for Data Analyst job position in Banglore Location on naukri.com (First 10 jobs data)

```python
#importing required libraries
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
```

```python
#connecting to the web-driver
driver=webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('https://www.naukri.com/')
```

```python
#finding element for job search bar and location
search_job=driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys("Data Analyst")
search_loc=driver.find_element_by_xpath("//input[@id='qsb-location-sugg']")
search_loc.send_keys("Bangalore")
```

```python
search_btn=driver.find_element_by_xpath("//div[@class='search-btn']/button")
search_btn.click()
```

```python
#Creating 4 empty lists
job_titles=[]
company_names=[]
location_list=[]
experience_list=[]
```

```python
#Extracting the tags having job-titles:

titles_tags=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
titles_tags
```

```python
for i in companies_tags:
    company_name=i.text
    company_names.append(company_name)
company_names[0:10]
```

```python
#Extracting all the tags having experience required data:

experience_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']/span")
experience_tags
```

```python
for i in experience_tags:
    experience=i.text
    experience_list.append(experience)
experience_list[0:10]
```

```python
#Extracting all the tags having location information:

location_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span")
location_tags
```

```python
for i in location_tags:
    location=i.text
    location_list.append(location)
location_list[0:10]
```

```python
#Making Dataframe:

#import pandas as pd
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
pd.set_option('display.width',None)
jobs=pd.DataFrame({})
jobs['Job-title']=job_titles[0:10]
jobs['Job-location']=location_list[0:10]
jobs['Company_name']=company_names[0:10]
jobs['Experience_required']=experience_list[0:10]
jobs
```

```python
driver.close()
```

# Answer-2) Scraping Data for Data Scientist job position in Banglore Location on naukri.com (First 10 jobs data)

```python
driver=webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('https://www.naukri.com/')
```

```python
#finding element for job search bar and location
search_job=driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys("Data Scientist")
search_loc=driver.find_element_by_xpath("//input[@id='qsb-location-sugg']")
search_loc.send_keys("Bangalore")
```

```python
search_btn=driver.find_element_by_xpath("//div[@class='search-btn']/button")
search_btn.click()
```

```python
#specifying the url of the webpage to be scrapepd.
url="https://www.naukri.com/data-scientist-jobs-in-bangalore-bengaluru?k=data%20scientist&l=bangalore%2Fbengaluru"

#opening the webpage through webdriver
driver.get(url)
```

```python
#Creating 4 empty lists
job_titles1=[]
company_names1=[]
location_list1=[]
job_description=[]
```

```python
#Extracting the tags having job-titles:

titles_tags1=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
titles_tags1

for i in titles_tags1:
    title=i.text
    job_titles1.append(title)
job_titles1[0:10]
```

```python
#extracting tags having comapny names :

companies_tags1=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
companies_tags1

for i in companies_tags1:
    company_name=i.text
    company_names1.append(company_name)
company_names1[0:10]
```

```python
#Extracting all the tags having location information:

location_tags1=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span")
location_tags1

for i in location_tags1:
    location=i.text
    location_list1.append(location)
location_list1[0:10]
```

```python
#For fetching Job description , we need to fetch each job title's url first
#fetching urls
import time
urls=[]
for i in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
    urls.append(i.get_attribute("href"))
urls
```

```python
#Extracting job-description

for i in urls:
    driver.get(i)
    time.sleep(5)
    
    try:
        job_desc=driver.find_element_by_xpath("//section[@class='job-desc']")
        
        job_description.append(job_desc.text)
    except:
        job_description.append('-')
job_description[0:10]
```

```python
#Making Dataframe:

#import pandas as pd
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
pd.set_option('display.width',None)
jobs1=pd.DataFrame({})
jobs1['Job-title']=job_titles1[0:10]
jobs1['Job-location']=location_list1[0:10]
jobs1['Company_name']=company_names1[0:10]
jobs1['Job-description']=job_description[0:10]
jobs1
```

```python
driver.close()
```

# Answer-5) Scraping the salary data for Data Scientist designation in Noida location on glassdoor's website(First 10 jobs data)

```python
driver.get('https://www.glassdoor.co.in/Salaries/index.htm')
```

```python
#finding element for job title search bar and location :
search_job=driver.find_element_by_id('KeywordSearch')
search_job.send_keys("Data Scientist")
search_loc=driver.find_element_by_id('LocationSearch')
search_loc.send_keys("Noida")
```

```python
#specifying the url of the webpage to be scrapped

driver.get("https://www.glassdoor.co.in/Salaries/new-delhi-data-scientist-salary-SRCH_IL.0,9_IM1083_KO10,24.htm?clickSource=searchBtn")
```

```python
#Creating 5 empty lists
company_names4=[]
no_of_salaries=[]
avg_salary=[]
min_salary1=[]
max_salary=[]
```

```python
#Extracting Company names

companies_tags4=driver.find_elements_by_xpath("//a[@class='css-f3vw95 e1aj7ssy3']")
companies_tags4

for i in companies_tags4:
    company_name=i.text
    company_names4.append(company_name)
company_names4[0:10]
```

```python
#Extracting no. of salaries

salaries=driver.find_elements_by_xpath("//span[@class='m-0 css-1b6bxoo']")

for i in salaries:
    salary=i.text
    no_of_salaries.append(salary)
no_of_salaries[2:12]
```

```python
#Extracting avg salary

avgsalary=driver.find_elements_by_xpath("//div[@class='col-12 col-lg-4 px-lg-0 d-flex align-items-baseline']")

for i in avgsalary:
    avsal=i.text.replace('\n','')
    avg_salary.append(avsal)
avg_salary[0:10]
```

```python
#Extracting min salary

minsalary=driver.find_elements_by_xpath("//div[@class='d-flex mt-xxsm css-79elbk epuxyqn0']/p[1]")

for i in minsalary:
    minsal=i.text
    min_salary1.append(minsal)
min_salary1[0:10]
```

```python
#Extracting max salary

maxsalary=driver.find_elements_by_xpath("//div[@class='d-flex mt-xxsm css-79elbk epuxyqn0']/p[2]")

for i in maxsalary:
    maxsal=i.text
    max_salary.append(maxsal)
max_salary[0:10]
```

```python
#Making Dataframe:

#import pandas as pd
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
pd.set_option('display.width',None)
jobs4=pd.DataFrame({})
jobs4['Company_name']=company_names4[0:10]
jobs4['Number of salaries']=no_of_salaries[2:12]
jobs4['Average salary']=avg_salary[0:10]
jobs4['Min salary']=min_salary1[0:10]
jobs4['Max salary']=max_salary[0:10]
jobs4
```

# Answer-10) Scraping 10 laptop data of intel core-i7 and intel core-i9 cpu type from amazon.com

Scraping title,ratings and price of the laptop

```python
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome('chromedriver.exe')
```

```python
driver.get("https://www.amazon.in/")
```

```python
#searching an element
search_laptop=driver.find_element_by_id('twotabsearchtextbox')
search_laptop.send_keys('Laptop')
```

```python
#clicking action
driver.find_element_by_id('nav-search-submit-button').click()
```

```python
#locating and clicking on core-i7 filter:

filter1=driver.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']/span")

for i in filter1:
    if i.text=='Intel Core i7':
        i.click()
        break
```

```python
#locating and clicking on core-i9 filter:

filter2=driver.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']/span")

for i in filter2:
    if i.text=='Intel Core i9':
        i.click()
        break
```

```python
#Specifying the url of the webpage to be scrapped:

driver.get("https://www.amazon.in/s?k=Laptop&i=computers&rh=n%3A1375424031%2Cp_n_feature_thirteen_browse-bin%3A12598163031%7C16757432031&dc&qid=1634429738&rnid=12598141031&ref=sr_nr_p_n_feature_thirteen_browse-bin_27")
```

```python
#fetching urls of each product on the page
urls=[]
for i in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):
    urls.append(i.get_attribute("href"))
urls
```

```python
#defining variables:
Title1=[]
Rating1=[]
Pricef1=[]
```

```python
for i in urls:
    driver.get(i)
    time.sleep(3)
    
    #fetching job title
    try:
        job=driver.find_element_by_xpath("//span[@id='productTitle']")
        Title1.append(job.text.replace("\n","new line"))
    except:
        Title1.append('-')
        
    #fetching rating
    try:
        ratingf=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base a-text-beside-button a-text-bold']")
        Rating1.append(ratingf.text)
    except:
        Rating1.append('-')
        
    #fetching price
    try:
        pricef=driver.find_element_by_xpath("//span[@id='priceblock_dealprice']")
        Pricef1.append(pricef.text)
    except:
        Pricef1.append('-')
```

```python
Title1[0:10]
```

```python
Rating1[0:10]
```

```python
Pricef1[0:10]
```

```python
#Making Dataframe:

#import pandas as pd
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
pd.set_option('display.max_colwidth',None)
laptop=pd.DataFrame({})
laptop['Title']=Title1[0:10]
laptop['Rating']=Rating1[0:10]
laptop['Price']=Pricef1[0:10]
laptop
```
